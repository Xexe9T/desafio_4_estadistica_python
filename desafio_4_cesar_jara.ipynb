Descripción
La base de datos world-data-2023.csv proporciona una gran cantidad de información sobre
todos los países del mundo, abarcando una amplia gama de indicadores y atributos. Incluye
estadísticas demográficas, indicadores económicos, factores ambientales, métricas de
atención médica, estadísticas educativas y mucho más. Con la representación de cada país,
este conjunto de datos ofrece una perspectiva global completa sobre diversos aspectos de
las naciones, lo que permite análisis en profundidad y comparaciones entre países. Las
variables consideradas son:
- 0. Country: Nombre del país.
- 1. Density (P/Km2): Densidad de población medida en personas por kilómetro
cuadrado.
- 2. Abbreviation: Abreviatura o código que representa el país.
- 3. Agricultural Land (%): Porcentaje del área de tierra utilizada para fines agrícolas.
- 4. Land Area (Km2): Área total de tierra del país en kilómetros cuadrados.
- 5. Armed Forces Size: Tamaño de las fuerzas armadas en el país.
- 6. Birth Rate: Número de nacimientos por 1,000 habitantes por año.
- 7. Calling Code: Código de llamada internacional para el país.
- 8. Capital/Major City: Nombre de la capital o ciudad principal.
- 9. CO2 Emissions: Emisiones de dióxido de carbono en toneladas.
- 10. CPI: Índice de Precios al Consumidor, una medida de la inflación y el poder
adquisitivo.
- 11. CPI Change (%): Cambio porcentual en el Índice de Precios al Consumidor en
comparación con el año anterior.
- 12. Currency_Code: Código de moneda utilizado en el país.
- 13. Fertility Rate: Número promedio de hijos nacidos de una mujer durante su vida.
- 14. Forested Area (%): Porcentaje del área de tierra cubierta por bosques.
- 15. Gasoline_Price: Precio de la gasolina por litro en moneda local.
- 16. GDP: Producto Interno Bruto, el valor total de bienes y servicios producidos en el
país.
- 17. Gross Primary Education Enrollment (%): Tasa de inscripción bruta en educación
primaria.
- 18. Gross Tertiary Education Enrollment (%): Tasa de inscripción bruta en educación
terciaria.
- 19. Infant Mortality: Número de muertes por cada 1,000 nacidos vivos antes de
cumplir un año de edad.
- 20. Largest City: Nombre de la ciudad más grande del país.
- 21. Life Expectancy: Número promedio de años que se espera que viva un recién
nacido.
- 22. Maternal Mortality Ratio: Número de muertes maternas por cada 100,000
nacidos vivos.
- 23. Minimum Wage: Nivel de salario mínimo en moneda local.
- 24. Official Language: Idioma(s) oficial(es) hablado(s) en el país.
- 25. Out of Pocket Health Expenditure (%): Porcentaje del gasto total en salud
pagado directamente por los individuos.
- 26. Physicians per Thousand: Número de médicos por cada mil personas.
- 27. Population: Población total del país.
- 28. Population: Labor Force Participation (%): Porcentaje de la población que forma
parte de la fuerza laboral.
- 29. Tax Revenue (%): Ingresos fiscales como porcentaje del PIB.
- 30. Total Tax Rate: Carga tributaria total como porcentaje de las ganancias
comerciales.
- 31. Unemployment Rate: Porcentaje de la fuerza laboral que está desempleada.
- 32. Urban Population: Porcentaje de la población que vive en áreas urbanas.
- 33. Latitude: Coordenada de latitud de la ubicación del país.
- 34. Longitude: Coordenada de longitud de la ubicación del país.

```python
# Importamos las librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargamos los datos creando un DataFrame llamado 'df'
df = pd.read_csv('world-data-2023.csv')
df.head()
```

**3. Considera los dos pares de variables anteriores y elimina en cada caso los faltantes**
**y atípicos. Justifica en cada caso tu procedimiento**

```python
# Crear el gráfico de dispersión entre variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Birth Rate', y='Fertility Rate', data=df_numerico)
plt.title('Relación entre Birth Rate y Fertility Rate')
plt.xlabel('Birth Rate (Tasa de Natalidad)')
plt.ylabel('Fertility Rate (Tasa de Fertilidad)')
plt.grid(True)
plt.show()
```

```python
# Eliminamos los datos faltantes
df_numerico.dropna(inplace=True)
```

```python
# Verificamos si hay nulos
df_numerico.isnull().sum()
```

```python
# Eliminacion de datos atipicos
# Cálculo del primer cuartil (Q1) y tercer cuartil (Q3) para la columna 'Birth Rate'
p25_br = df_numerico['Birth Rate'].quantile(0.25)  # Q1: percentil 25
p75_br = df_numerico['Birth Rate'].quantile(0.75)  # Q3: percentil 75

# Cálculo del rango intercuartílico (IQR)
rango_iqr_br = (p75_br - p25_br)  # IQR: diferencia entre Q3 y Q1

# Cálculo de los límites inferior y superior para identificar valores atípicos
limite_inferior_br = p25_br - 1.5 * rango_iqr_br  # Límite inferior: Q1 - 1.5 * IQR
limite_superior_br = p75_br + 1.5 * rango_iqr_br  # Límite superior: Q3 + 1.5 * IQR

# Cálculo del primer cuartil (Q1) y tercer cuartil (Q3) para la columna 'Fertility Rate'
p25_fr = df_numerico['Fertility Rate'].quantile(0.25)  # Q1: percentil 25
p75_fr = df_numerico['Fertility Rate'].quantile(0.75)  # Q3: percentil 75

# Cálculo del rango intercuartílico (IQR)
rango_iqr_fr = (p75_fr - p25_fr)  # IQR: diferencia entre Q3 y Q1

# Cálculo de los límites inferior y superior para identificar valores atípicos
limite_inferior_fr = p25_fr - 1.5 * rango_iqr_fr  # Límite inferior: Q1 - 1.5 * IQR
limite_superior_fr = p75_fr + 1.5 * rango_iqr_fr  # Límite superior: Q3 + 1.5 * IQR

# Filtrado del DataFrame para excluir los valores atípicos en 'Fertility Rate'
# Filtrar el DataFrame para eliminar los valores atípicos de 'Birth Rate' y 'Fertility Rate'
df_fertilidad_nacimiento = df_numerico[
    (df_numerico['Birth Rate'].between(limite_inferior_br, limite_superior_br)) &
    (df_numerico['Fertility Rate'].between(limite_inferior_fr, limite_superior_fr))
]

# Mostrar el DataFrame filtrado
df_fertilidad_nacimiento
```

```python
# Crear el gráfico de dispersión entre variables ahora sin datos faltantes y sin datos atipicos
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Birth Rate', y='Fertility Rate', data=df_fertilidad_nacimiento)
plt.title('Relación entre Birth Rate y Fertility Rate (Sin Faltantes, ni Atipicos)')
plt.xlabel('Birth Rate (Tasa de Natalidad)')
plt.ylabel('Fertility Rate (Tasa de Fertilidad)')
plt.grid(True)
plt.show()
```

2. Impacto de los Atípicos en el Análisis:

Importancia de Limpiar los Datos:

En resumen, limpiar los datos es fundamental para realizar un análisis de calidad y obtener resultados confiables.


### Conclusiones del Gráfico:

1. **Grandes Diferencias en el Tamaño de las Economías**:
   - El gráfico muestra que hay enormes diferencias en el tamaño de las economías de los países. Estados Unidos y China están muy alejados del resto porque tienen un Producto Interno Bruto (GDP) mucho más alto que cualquier otro país. Esto significa que son las economías más grandes del mundo.

2. **Relación entre Población y GDP**:
   - Hay una tendencia general a que los países con más población tengan un GDP más alto. Por ejemplo, China e India, los países más poblados del mundo, también tienen un GDP significativo. Sin embargo, esto no es una regla absoluta, ya que hay países como Japón y Alemania que, con una población más pequeña, tienen un GDP alto debido a sus economías desarrolladas y productivas.

3. **Muchos Países con GDP Bajo**:
   - La mayoría de los países están agrupados en la parte inferior izquierda del gráfico, lo que indica que muchos países tienen un GDP bajo y también poblaciones más pequeñas o medianas. Esto significa que la mayoría de los países no tienen economías tan grandes.

4. **Países con Economías Fuertes pero Poblaciones Menores**:
   - Algunos países, como Japón y Alemania, tienen un GDP alto a pesar de tener una población más pequeña en comparación con países como India o China. Esto sugiere que estos países son muy productivos y tienen una economía fuerte en relación con su tamaño de población.

### ¿Qué Nos Dice Esto?

- Este gráfico nos ayuda a ver cómo se relacionan la población y el tamaño de la economía de los países. Aunque tener una gran población puede contribuir a un mayor GDP, no es el único factor. La eficiencia económica, la tecnología y otros recursos también son importantes para determinar cuán grande es la economía de un país.

En resumen, aunque hay cierta correlación entre población y GDP, la fuerza de una economía también depende de otros factores más allá del tamaño de la población.

**7. Escoge otro par de datos, límpialos, prepáralos y analiza su posible relación. Justifica**
**utilizando las herramientas vistas y gráficos adecuados.**

```python
# Usaremos las variables 'Infant mortality' y 'Life expectancy'

# Cálculo del primer cuartil (Q1) y tercer cuartil (Q3) para la columna 'Infant mortality'
p25_im = df_numerico['Infant mortality'].quantile(0.25)  # Q1: percentil 25
p75_im = df_numerico['Infant mortality'].quantile(0.75)  # Q3: percentil 75

# Cálculo del rango intercuartílico (IQR) para 'Infant mortality'
rango_iqr_im = (p75_im - p25_im)  # IQR: diferencia entre Q3 y Q1

# Cálculo de los límites inferior y superior para identificar valores atípicos en 'Infant mortality'
limite_inferior_im = p25_im - 1.5 * rango_iqr_im  # Límite inferior: Q1 - 1.5 * IQR
limite_superior_im = p75_im + 1.5 * rango_iqr_im  # Límite superior: Q3 + 1.5 * IQR

# Cálculo del primer cuartil (Q1) y tercer cuartil (Q3) para la columna 'Life expectancy'
p25_le = df_numerico['Life expectancy'].quantile(0.25)  # Q1: percentil 25
p75_le = df_numerico['Life expectancy'].quantile(0.75)  # Q3: percentil 75

# Cálculo del rango intercuartílico (IQR) para 'Life expectancy'
rango_iqr_le = (p75_le - p25_le)  # IQR: diferencia entre Q3 y Q1

# Cálculo de los límites inferior y superior para identificar valores atípicos en 'Life expectancy'
limite_inferior_le = p25_le - 1.5 * rango_iqr_le  # Límite inferior: Q1 - 1.5 * IQR
limite_superior_le = p75_le + 1.5 * rango_iqr_le  # Límite superior: Q3 + 1.5 * IQR

# Filtrado del DataFrame para excluir los valores atípicos en 'Infant mortality' y 'Life expectancy'
df_mortalidad_expectativa = df_numerico[
    (df_numerico['Infant mortality'].between(limite_inferior_im, limite_superior_im)) &
    (df_numerico['Life expectancy'].between(limite_inferior_le, limite_superior_le))
]

# Graficar comparativamente los datos originales y los datos sin outliers
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 1: Datos originales
sns.scatterplot(ax=axes[0], x='Infant mortality', y='Life expectancy', data=df_numerico)
axes[0].set_title('Relación entre Infant Mortality y Life Expectancy (Original)')
axes[0].set_xlabel('Infant Mortality (Mortalidad Infantil)')
axes[0].set_ylabel('Life Expectancy (Esperanza de Vida)')
axes[0].grid(True)

# Gráfico 2: Datos sin outliers
sns.scatterplot(ax=axes[1], x='Infant mortality', y='Life expectancy', data=df_mortalidad_expectativa)
axes[1].set_title('Relación entre Infant Mortality y Life Expectancy (Sin Outliers)')
axes[1].set_xlabel('Infant Mortality (Mortalidad Infantil)')
axes[1].set_ylabel('Life Expectancy (Esperanza de Vida)')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

### Conclusión:

Los gráficos muestran una clara relación negativa entre la mortalidad infantil y la esperanza de vida. Esto significa que, en general, a medida que la mortalidad infantil es más baja, la esperanza de vida tiende a ser más alta. En el primer gráfico (datos originales), se pueden ver algunos puntos que no siguen esta tendencia general, lo que puede deberse a valores atípicos. 

En el segundo gráfico (sin valores atípicos), la relación entre las dos variables se vuelve más clara y consistente. Al eliminar los datos extremos, podemos ver mejor la tendencia principal: los países con menores tasas de mortalidad infantil suelen tener una mayor esperanza de vida. Esto refuerza la idea de que mejorar la salud infantil contribuye a una mayor longevidad en la población.
